```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "arrhythmia.data",
    header=None,
    na_values='?'
)
```

```python
df.head(5)
```

```python
df.info()
```

```python
# check the percentage
df.isna().sum() / len(df)
```

```python
df.isnull().sum().sum()
```

```python
missing_per_col = df.isnull().sum()
missing_per_col[missing_per_col > 0]
```

```python
# Visualizing Missing Data using a seaborn heatmap.
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
```

```python
df.columns = [f'feature_{i}' for i in range(df.shape[1])]
```

```python
df.columns[:5]
```

```python
col = df['feature_14']

Q1 = col.quantile(0.25)
Q3 = col.quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = col[(col < lower) | (col > upper)]
len(outliers)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.boxplot(x=df['feature_14'])
plt.title("Outlier Detection – feature_14")
plt.show()
```
